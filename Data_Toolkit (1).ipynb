# Data Toolkit


**1. What is NumPy, and why is it widely used in Python?**  
- NumPy is a core library for numerical computing in Python. It provides powerful tools for handling large, multi-dimensional arrays and matrices, along with a suite of high-level mathematical functions. NumPy is widely used due to its performance efficiency, vectorized operations, and ability to seamlessly integrate with other scientific libraries.

**2. How does broadcasting work in NumPy?**  
- Broadcasting in NumPy is a technique that allows operations between arrays of different shapes. Instead of raising errors, NumPy intelligently expands smaller arrays across dimensions to match the shape of larger ones. This makes arithmetic operations more flexible and efficient, particularly when working with scalars or lower-dimensional arrays.

**3. What is a Pandas DataFrame?**  
- A Pandas DataFrame is a two-dimensional labeled data structure, akin to a spreadsheet or SQL table. It allows for storage and manipulation of tabular data with ease. Each column can hold different data types, and the structure supports various functions like filtering, aggregation, and reshaping.

**4. Explain the use of the `groupby()` method in Pandas.**  
- The `groupby()` method in Pandas is used to split data into groups based on the values of one or more columns. It is most commonly followed by aggregation functions such as `sum()`, `mean()`, or `count()`, allowing analysts to derive insights by examining grouped patterns and summaries.

**5. Why is Seaborn preferred for statistical visualizations?**  
- Seaborn is favored for statistical visualizations because it simplifies complex plotting routines with minimal code. Built on Matplotlib, it offers elegant themes, clear syntax, and automatic statistical estimation features. This makes it perfect for plots like distribution charts, heatmaps, and regression visualizations.

**6. What are the differences between NumPy arrays and Python lists?**  
- NumPy arrays differ from Python lists primarily in performance and functionality. Arrays are homogeneously typed and optimized for numerical calculations, offering faster and more memory-efficient processing. Lists, on the other hand, can store mixed types but are slower and lack built-in mathematical operations.

**7. What is a heatmap, and when should it be used?**  
- A heatmap is a graphical representation where data values are depicted using color gradients. It’s particularly useful for visualizing correlations, identifying missing values, or highlighting intensity in datasets like confusion matrices or population densities.

**8. What does the term "vectorized operation" mean in NumPy?**  
- Vectorized operations refer to the execution of mathematical functions directly on whole arrays, rather than through iterative loops. This leads to cleaner code and substantial performance gains, as operations are internally optimized and compiled.

**9. How does Matplotlib differ from Plotly?**  
- Matplotlib is a traditional library designed for static, high-quality graphs suitable for research and publication. In contrast, Plotly is interactive and designed for web applications and dashboards. While Matplotlib offers deep customization, Plotly emphasizes ease of use and user interaction.

**10. What is the significance of hierarchical indexing in Pandas?**  
- Hierarchical indexing lets you create multiple levels of index values for rows or columns, enabling complex data representation and slicing. This is particularly helpful when working with multi-dimensional categorical data or time series nested by groups.

**11. What is the role of Seaborn's `pairplot()` function?**  
- The `pairplot()` function in Seaborn creates pairwise scatterplots for all numerical columns and shows distributions along the diagonal. It’s a quick and informative tool to assess relationships, detect trends, or spot outliers across multiple variables.

**12. What is the purpose of the `describe()` function in Pandas?**  
- The `describe()` function provides summary statistics for each numeric column in a DataFrame. This includes count, mean, standard deviation, min, max, and quartile values. It’s ideal for getting a quick snapshot of the data’s distribution.

**13. Why is handling missing data important in Pandas?**  
- Handling missing data is essential to preserve the integrity of analysis and modeling. Unaddressed missing values can skew results or cause errors. Pandas provides methods like `dropna()` and `fillna()` to remove or impute missing entries appropriately.

**14. What are the benefits of using Plotly for data visualization?**  
- Plotly excels in creating rich, interactive charts that users can zoom, pan, and hover over. It supports advanced visual types like 3D plots and integrates smoothly with web frameworks, making it a preferred choice for dashboards and analytics platforms.

**15. How does NumPy handle multidimensional arrays?**  
- NumPy uses the `ndarray` object to represent arrays of any dimension. These can be indexed and reshaped easily, and high-level functions allow seamless operations across dimensions, making it ideal for scientific simulations and tensor computations.

**16. What is the role of Bokeh in data visualization?**  
- Bokeh is a Python library that focuses on browser-based, interactive visualizations. It allows dynamic behaviors such as zooming and filtering, supports widgets and callbacks, and is suitable for building web dashboards directly from Python code.

**17. Explain the difference between `apply()` and `map()` in Pandas.**  
- The `map()` function applies a function element-wise to a Series, making it suitable for simple transformations. In contrast, `apply()` can be used on Series and DataFrames, and supports more complex row- or column-wise operations with custom logic.

**18. What are some advanced features of NumPy?**  
- NumPy offers a range of advanced features including broadcasting, masked arrays, structured arrays, universal functions (ufuncs), and memory-mapped files. These tools make it versatile for high-performance scientific and engineering applications.

**19. How does Pandas simplify time series analysis?**  
- Pandas simplifies time series tasks by offering datetime indexing, resampling methods, rolling statistics, and time zone handling. It allows time-based grouping, shifting, and interpolation, making it powerful for financial or temporal data modeling.

**20. What is the role of a pivot table in Pandas?**  
- A pivot table in Pandas rearranges data to highlight relationships and comparisons across categories. It’s an efficient way to compute grouped summaries and perform cross-tabulations, much like in spreadsheet applications.

**21. Why is NumPy's array slicing faster than Python's list slicing?**  
- NumPy slicing is faster because arrays are stored in contiguous memory and processed using optimized low-level C routines. Python lists, with their flexibility and dynamic typing, incur more overhead and lack such performance optimization.

**22. What are some common use cases for Seaborn?**  
- Seaborn is commonly used for visualizing distributions, correlations, and categorical comparisons. It supports plots like histograms, KDEs, box plots, violin plots, regression plots, and heatmaps, making it invaluable for exploratory data analysis.



```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np
arr = np.array([[1, 2], [3, 4], [5, 6]])
row_sums = np.sum(arr, axis=1)
print(row_sums)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd
df = pd.DataFrame({'Score': [90, 80, 70]})
mean_score = df['Score'].mean()
print(mean_score)
```

```python
# 3.Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [4, 5, 6]
plt.scatter(x, y)
plt.show()
```

```python
# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
data = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6], 'C':[7,8,9]})
corr = data.corr()
sns.heatmap(corr, annot=True)
```

```python
# 7.Write a program to perform element-wise multiplication of two NumPy arrays.

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
# 8.Create a line plot with multiple lines using Matplotlib.

plt.plot([1, 2, 3], [4, 5, 6], label='Line A')
plt.plot([1, 2, 3], [7, 8, 9], label='Line B')
plt.legend()
plt.show()
```

```python
# 9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 62, 90, 70]
}
df = pd.DataFrame(data)


threshold = 75

filtered_df = df[df['Score'] > threshold]
print(filtered_df)
```

```python
# 10.Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


data = pd.DataFrame({
    'Marks': [45, 55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
})


sns.histplot(data=data, x='Marks', bins=5, kde=True)


plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# 11.Perform matrix multiplication using NumPy.

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```
